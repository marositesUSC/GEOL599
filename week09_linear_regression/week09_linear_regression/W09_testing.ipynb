```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
t = np.arange(-1,1,0.1)
w1 = np.sin(t)
w2 = np.cos(t)
w3 = w1 + w2
```

```python
plt.plot(t, w1, label='sin wave')
plt.plot(t, w2, label='cos wave')
plt.plot(t, w3, label='sum wave')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

L = 10 # time interval, 1 second
t = np.linspace(-L,L,400) # time, sampled 200 times per second

# Sine wave 1
amp1 = 2
k1 = 1 # number of cycles/second
sin1=amp1*np.sin(2*np.pi*k1*t/L)

# Sine wave 2
amp2 = 1
k2 = 5 # number of cycles/second
sin2=amp2*np.sin(2*np.pi*k2*t/L)

# f(t)
f = sin1 + sin2

#Plot
plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(t,sin1)
plt.ylim([-3, 3])
plt.grid()
plt.xlabel('t, seconds')
plt.title('Sine Wave 1')
plt.subplot(1,3,2)
plt.plot(t,sin2)
plt.ylim([-3, 3])
plt.grid()
plt.xlabel('t, seconds')
plt.title('Sine Wave 2')
plt.subplot(1,3,3)
plt.plot(t,f)
plt.ylim([-3, 3])
plt.grid()
plt.xlabel('t, seconds')
plt.title('f(t) = Sum of Sine Waves 1 + 2')
plt.show()
```
