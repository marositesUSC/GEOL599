```python

for s in allStorms['data']['storms']:
    if "HAGIBIS" in s['name']:
        print(s)
        #trackUrl = baseUrl+s['stormID']
        #r = requests.get(trackUrl)
        #track = r.json()
        with open(f"/srv/data/hurricane/storm_tracks/{s['stormID']}.json", "r") as injson:
            track = json.load(injson)
            
```

```python
list(track['data']['storm']['features'][0])
```

```python
trackPts = []
trackLats = []
trackLons = []
for f in track['data']['storm']['features']:
    trackLons.append(f['geometry']["coordinates"][0])
    trackLats.append(f['geometry']["coordinates"][1])
                     
```

```python
plt.figure(figsize=(15,5))                                # Creates the plot *object* and sets the size
plt.rcParams['figure.dpi'] = 500                          # Sets the plot resolution to 500 Dots Per Inch

ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=170))              # Creates the axis *object* with a PlateCarree map projection (more on this later)
ax.stock_img()                                            # Uses the *method* stock_img
ax.set_title('Map of the World with Columbia')            # Gives the map a title
plt.scatter(trackLons, trackLats, transform=ccrs.PlateCarree(), color='red') # Uses the *scatter* to plot a symbol at the given lat and lon
plt.show()        #Writes out the figure to a png file
```
