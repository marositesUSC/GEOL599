# Variables in python

```python
a=3
print(a)
```

Think of a variable as a "box" with a name, that can hold onto a value.

They have **classes**. The **class** controls what it can do.

We can easily find the **class** of each variable using **type** function.

```python
type(a)
```

This tell us that variable ``a`` is of the class ``int``, or **integer**.

# integer (int)
Whole numbers only.

```python
age = 3
age
```

```python
print(type(age))
```

Notice how we have **nested functions**. We have the **function** ``type`` acting on the ``variable`` age. 

We then have the **function** ``print`` acting on the **result** of the **function** ``type`` acting on the ``variable`` age. 

Keep track of the number of parentheses used.

# boolean (bool)
Simply a value of True or False.

```python
something_is = True
something_is
```

```python
type(something_is)
```

Booleans are ***incredibly*** useful and we will come back to these a lot throughout the semester.

# float
Floats are **floating point** numbers, or decimals to you and me.

```python
temperature = 98.6
temperature
```

```python
print(type(temperature))
```

# string (str)
Strings are collections of characters. They are defined by being enclosed by quotations ``"`` or ``'``.

```python
name = "Bond, James Bond"
name
```

```python
print(type(name))
```

```python
name2 = 'Indiana Jones'
name2
```

**Strings** can have any value, including the contents to an **integer**, a **float**, a **boolean**, or anything else.

However, because it's a **string**, it works differently than other **classes** do.

```python
print (a+3)
```

```python
print (temperature * 2)
```

```python
temp_string = "98.6"
```

```python
print (temp_string / 3)
```

This causes an error.

Division isn't allowed for strings.

# list
A list of other things.

```python
things = [ 1, 2, 3]
print(things[1])
print(type(things))
```

The contents of a **list** can be of different **classes**.

```python
others = ["a", 2, 5.4]
print (others)
```

A property of lists is that we can change or **reassign** the values.

```python
things[0]=3
print (things)
```

# tuple

```python
couple_items = ( "a", "boat" , 4)
print(couple_items)
print(couple_items[1])
```

```python
print(type(couple_items))
```

A property of **tuples** is that we cannot change or **reassign** the values.

```python
couple_items[0] = "c"
```
