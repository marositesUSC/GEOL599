```python
import mpl_ternary as tn
```

```python
fig, ax = tn.ternary(upper_label='upper label', left_label='left label', right_label='right label')

ax.plot(tn.tri2cartX(1, 0, 0), tn.tri2cartY(1, 0, 0), 'o', markersize=12, label='(1,0,0)')
ax.plot(tn.tri2cartX(0, 1, 0), tn.tri2cartY(0, 1, 0), 'o', markersize=12, label='(0,1,0)')
ax.plot(tn.tri2cartX(0, 0, 1), tn.tri2cartY(0, 0, 1), 'o', markersize=12, label='(0,0,1)')
ax.plot(tn.tri2cartX(1/3, 1/3, 1/3), tn.tri2cartY(1/3, 1/3, 1/3), 'o', markersize=12, label='the centre')

ax.legend(fontsize=14)
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
```

```python
igneous_data = pd.read_csv('data/ign.csv',sep='\t',low_memory=False)
```

```python
volcanic_data = igneous_data[igneous_data['TYPE']=='VOLCANIC']
basalt=volcanic_data[volcanic_data['ROCK NAME']=='BASALT']
volcanic_data.head()
#SIO2	TIO2	AL2O3	FE2O3
```

```python
#basalt.iloc[0]['SIO2','FE2O3']
basalt.loc[0:1, ['SIO2','FE2O3','AL2O3']]
```

```python
fig, ax = tn.ternary(upper_label='SIO2', left_label='AL2O3', right_label='FE2O3')
#basalt['SIO2'],basalt['FE2O3'],basalt['AL2O3']
ax.plot(tn.tri2cartX(basalt['SIO2'],basalt['FE2O3'],basalt['AL2O3']), tn.tri2cartY(basalt['SIO2'],basalt['FE2O3'],basalt['AL2O3']), 'o', markersize=12, label='basalt')
ax.legend(fontsize=14)
```
