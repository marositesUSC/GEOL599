## Visualizing the data distribution

Let's use ```matplotlib``` to visual the distribution of elevations on Earth. The elevations are within the `topo2d_grid` array. 

First, we need to manipulate the data a bit. We want to have a 1-dimensional list of all the (540 x 1081 = 583740) (583740) element array values to plot the distribution of elevations. We can use the numpy function ```reshape``` to do this. Instead of a grid, the data will all be in a single row after it is reshaped. To do so, we are applying a function (in this case ```reshape```) to an object (in this case the topo2d_grid object) and saving the result as a new numpy array which we will call topo_array.

````python
print(topo2d_grid.shape)
topo_1d_len = topo2d_grid.shape[0]*topo2d_grid.shape[1]
print(topo_1d_len)
topo_array = topo2d_grid.reshape(topo_1d_len)
````

## Making a topographic map

Histograms and CDFs are two useful 1D visualizations of our data, but what about the distribution of topography across the globe? 

Again, we'll need to manipulate the data a bit. We need grids of latitude and longitude in order to map the topography. Like `topo2d_grid` they will be (size 540 x 1081). `lat2d_grid` will have rows of constant latitude, and `lon2d_grid` will have columns of constant longtiude. To do this we'll call the numpy function `meshgrid`.

<img src="lon_lat_grids.png">

````python
np.meshgrid?
````

`meshgrid` will repeat the lats and lons to create a mesh.

````python
lon2d_grid, lat2d_grid = np.meshgrid(lons, lats)
````

We can use `shape` to see the result.

````python
print(np.shape(lon2d_grid))
print(np.shape(lat2d_grid))
print(np.shape(topo2d_grid))
````

### Import cartopy

We are going to use the module `cartopy` to make maps using its coordinate reference tools so let us import it.

````python
import cartopy.crs as ccrs
````

### Use the `contourf` function to make the map

Use the grids of latitude and longtitude to plot a Robinson projection map of topography using the `cartopy` function `contourf`. This function is *slow* so do not worry if it takes a while to run. First, what does `contourf` do?

````python
plt.contour?
````

Now let's plot the map.

````python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection=ccrs.Robinson())
ax.set_global()
plt.contourf(lon2d_grid, lat2d_grid, topo2d_grid, levels=10, transform=ccrs.PlateCarree())

ax.coastlines()
ax.gridlines()

plt.colorbar(ax=ax,orientation='horizontal',label='elevation (m)')

plt.show()
````

### Visualizing the shallowest and deepest parts of the ocean

Let us use the `Orthographic` projection (A GLOBE!) to look at the large underwater mountain range of the Mid-Atlantic ridge. 

<font color=goldenrod>**_Code for you to write_**</font>

The code block below needs the `plt.contourf` function to be completed. Use the version of the inputs that we used above as a guide. In addition, add the `vmax` parameter (which means that all values above this maximum will have the same colour) and set it equal to zero so that the colorbar only corresponds to elevations below sea level.

````python
fig = plt.figure(figsize=(15,10))
ax = plt.axes(projection=ccrs.Orthographic(central_longitude=-40, central_latitude=20))
ax.set_global()
plt.contourf(lon2d_grid, lat2d_grid, topo2d_grid, levels=20, transform=ccrs.PlateCarree(), vmax=0)

ax.coastlines()
ax.gridlines()

plt.colorbar(ax=ax,orientation='horizontal',label='elevation (m)')

plt.show()
````
