```python
#Import in the necessary modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import signal
from scipy.interpolate import interp1d
```

Read in the data from `./data/LR04stack.csv`

```python
#Read in the data and view it with .tail()
LR04_d18O = pd.read_csv('data/LR04stack.csv')
LR04_d18O.tail()
```

```python
#Plot $\delta ^{18}$O (y-axis) versus time (x-axis) for the whole dataset [1 point]
plt.figure(figsize=(16,8))
plt.plot(LR04_d18O['Age (ka)'], LR04_d18O['d18O'], label=r'$\delta ^{18}$')
plt.title('benthic oxygen isotope ($\\delta ^{18}$) Data from Lisecki and Raymo (2005) \n5.32 million years to Present')
plt.xlim(LR04_d18O['Age (ka)'].max(), 0)
plt.xlabel('Time (kya)')
plt.ylabel(r'$\delta ^{18}$')
plt.grid()
plt.show()
```

```python
### COPIED FROM IN CLASS ASSIGNMENT. 

#Interpolate the data [1 point]
d18O_interp = interp1d(LR04_d18O['Age (ka)'], LR04_d18O['d18O'], kind='linear')

#Crate the uniform time series
age_uniform = np.linspace(0, 5320,5321,True)

#Create a new interpolated dataframe
#Interpolate the d18O values to the uniform time series
d18O_interp = d18O_interp(age_uniform)

#Create a new interpolated dataframe
interpolated_dataframe={'Age (ka)': age_uniform, 'd18O': d18O_interp}
LR04_d18O_interp = pd.DataFrame(data=interpolated_dataframe)
```

```python
#Plot the interpolated and original data on the same plot. Plot interpolated data as a line and the original data as points.
plt.scatter(LR04_d18O['Age (ka)'],LR04_d18O['d18O'], color='darkblue', s=8, zorder=1, label='LR04 Data Set')
plt.plot(LR04_d18O_interp['Age (ka)'],LR04_d18O_interp['d18O'], color='grey', zorder=0, label='Linear Interpolation')


plt.title('LR04 oxygen isotopic variation stack', fontsize=14)
plt.xlabel('Age (ka)', fontsize=14)
plt.ylabel('$\\delta ^{18}$O', fontsize=14)
# plt.ylim(5,3)
plt.xlim(1000,0)
plt.xticks(range(0,1000,100))
plt.legend()
plt.grid()
plt.show()
```

```python
#Filter the interpolated data into the following intervals
#1. 0-950ka
#2. 950-1900ka  -- Calabrian-ish 
#3. 1900-2850ka -- Gelasian-ish 
#4. 2850-5320ka -- Piacenzian-ish 

#Make 4 separate dataframes, one for each time interval
d18O_0_950 = LR04_d18O_interp[(LR04_d18O_interp['Age (ka)']<=950)]   
d18O_950_1900 = LR04_d18O_interp[(LR04_d18O_interp['Age (ka)']>950) & (LR04_d18O_interp['Age (ka)']<=1900)]
d18O_1900_2850 = LR04_d18O_interp[(LR04_d18O_interp['Age (ka)']>1900) & (LR04_d18O_interp['Age (ka)']<=2850)]
d18O_2850_5320 = LR04_d18O_interp[(LR04_d18O_interp['Age (ka)']>2850)]   #  & (LR04_d18O_interp['Age (ka)']<5320)

print(len(d18O_0_950), len(d18O_950_1900), len(d18O_1900_2850), len(d18O_2850_5320))
```

### Interval 1. between 0-950 ka [2 points]

```python
# Plot $\delta ^{18}$O as a function of time for 0-950 ka

plt.plot(d18O_0_950['Age (ka)'],d18O_0_950['d18O'])
plt.title('LR04 oxygen isotopic variation \n950 kya - Present', fontsize=14)
plt.xlabel('Age (ka)', fontsize=14)
plt.ylabel('$\\delta ^{18}$O', fontsize=14)
plt.ylim(5,3)
plt.xlim(d18O_0_950['Age (ka)'].max(),d18O_0_950['Age (ka)'].min())
plt.xticks(range(0,1000,100))
plt.show()
```

```python
eccentricity_frequency = 1.0/100.0
obliquity_frequency = 1.0/41.0
precession_frequency_1 = 1.0/23.0
precession_frequency_2 = 1.0/19.0
```

```python
#Calculate the periodogram
frequency_d18O_0_950, power_d18O_0_950 = signal.periodogram(d18O_0_950['d18O'], 1) 
```

```python
#Plot two subplots of power versus frequency and then power versus frequency on a log-y plot 
plt.subplots(2,1,figsize=(16,6),sharex=True)

plt.subplot(2,1,1)
plt.plot(frequency_d18O_0_950, power_d18O_0_950, label='Power vs. Frequency')
plt.title('Power Spectra')
plt.ylabel('Power')
plt.xlim(0,0.1)
plt.grid()
plt.axvline(x=eccentricity_frequency,color='red',label='Eccentricity',linewidth=2,linestyle='--') 
plt.axvline(x=obliquity_frequency,color='orange',label='Obliquity',linewidth=2,linestyle='--')
plt.axvline(x=precession_frequency_1,color='gold',label='Precession',linewidth=2,linestyle='--') 
plt.axvline(x=precession_frequency_2,color='green',label='Precession',linewidth=2,linestyle='--')

plt.subplot(2,1,2)
plt.semilogy(frequency_d18O_0_950, power_d18O_0_950, label='Power vs. Frequency')
plt.title('Power Spectra on log-y plot')
plt.xlabel('Frequency 1/kya')
plt.ylabel('Power')
plt.ylim(10e-3,10e1)
plt.grid()
plt.axvline(x=eccentricity_frequency,color='red',label='Eccentricity',linewidth=2,linestyle='--') 
plt.axvline(x=obliquity_frequency,color='orange',label='Obliquity',linewidth=2,linestyle='--')
plt.axvline(x=precession_frequency_1,color='gold',label='Precession',linewidth=2,linestyle='--') 
plt.axvline(x=precession_frequency_2,color='green',label='Precession',linewidth=2,linestyle='--')

plt.tight_layout()
plt.show()

```

## Interpretation

After you have done your analysis, you will find that there are different periods for the glacial-interglacial cycles in the past 0-950 ka as opposed to those >950 ka ago.

**What is the period of the more recent glacial-interglacial cycles? What orbital forcing might it correspond to? [2 points]**

Recent glacial-interglacial periods occurred most prominently in 100k-year cycles. This aligns most closely with solar forcing tied to the shape of Earth's orbit, or Eccentricity. To a lesser degree, Obliquity, or the tilt of Earth's axis, has also played a role. 

**What is the period of the more ancient glacial-interglacial cycles? What orbital forcing might it correspond to? [2 points]**

More ancient glacial-interglacial periods occured more along a 40k year cycle indicating obliquity was a major driver of past cycles.

**This change in glacial-interglacial cycle period is often called the mid-Pleistocene transition. Using Google Scholar, find a paper that presents a hypothesis for what caused the transition. Provide the reference here and write a paragraph summarizing the hypothesis. Use your own critical thinking skills, not ChatGPT's [4 points]**

The prevailing consensus is that the mid-Pleistocene transition (MPT) is the result of nonlinear internal dynamics that have shifted the Earth's climate system from an obliquity-centric state (40k) to a more eccentricity-driven state (100k). There is some dispute about the specific drivers of the transition because of the expected weakness of solar forcing from eccentricity, which has led to the ***100k problem***. Huybers' ([2007](https://doi.org/10.1016/j.quascirev.2006.07.013)) work summarized it nicely and explored it using a depth-derived age model using the oxygen-isotope record. This model confirmed that there must be some driver in the climate system and explained the impacts of a suspected decrease in atmospheric $CO_2$ or the removal of *slippery* regolith, both of which would lead to thicker, more persistent ice sheets and a more stable climate system. Huybers' work goes on to challenge several leading theories, specifying that eccentricity is not the main driver of the climate system over the Late Pleistocene, but that obliquity remains the leading driver, requiring several cycles to reach a critical threshold and transition between glacial and interglacial periods on an 80-120k year cycle. 


Huybers, Peter. 2007. Glacial Variability over the Last Two Million Years: An Extended Depth-Derived Agemodel, Continuous Obliquity Pacing, and the Pleistocene Progression. Quaternary Science Reviews 26 (12): 3755. https://doi.org/10.1016/j.quascirev.2006.07.013.
