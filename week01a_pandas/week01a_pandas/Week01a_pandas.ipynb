### Get rows via boolean expression

We can select data using **boolean** expressions. These evaluate if something is **True** or **False**. The expression below is **True** where the **Year** is greater than or equal to 2010, and **False** everywhere else.

```python
solPrice['Year'] >= 2010
```

So how can this help? We can use this to **mask** or select data. We use the **boolean values** as an **index** using square brackets `[ ]`.

Look at how applying this **mask** changes the result.

```python
solPrice[ (solPrice['Year'] >= 2010) ~ (solPrice['Year'] = 2015) ]
```

But notice that this hasn't changed the contents of the original dataframe, since we didn't save it.

```python
print (solPrice)
```

## Other subsets using booleans
We can also **mask** a dataframe by find where values equal something.

For example, if we wanted to find where the data for the year 2020, we would first create a **mask** that was **True** for only Year=2020

```python
solPrice['Year'] == 2020
```

<font color=goldenrod>**_Code for you to write_**</font>

Apply this mask to the whole dataframe and show the ``Solar photovoltaic cumulative capacity`` in that year.

We may want to know the **index** of the array (i.e. row number) for the Year of 2020. To do that we display the index using the ``.index`` function, and only show the first value.

- Get the Series for Year
- Subset for only year 2020
- Get the index as a Series
- Get the first (0th) item in the Series

```python
row_for_2020 = solPrice[solPrice['Year'] == 2020].index[0]  # gets index of first r45
print(f"Row for 2020 is {row_for_2020}")
```
