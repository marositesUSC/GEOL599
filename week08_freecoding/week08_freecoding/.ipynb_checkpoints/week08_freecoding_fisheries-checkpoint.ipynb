# Cod collapse
In this unit we will be using the RAM Legacy Database: 

https://www.ramlegacy.org/

The RAM Legacy Stock Assessment Database is a compilation of stock assessment results for commercially exploited marine populations from around the world. It is inspired by Dr. Ransom A. Myers’ original stock-recruitment database, which is no longer being updated.

Let's examine the infamous collapse of the Canadian cod stock. 

Newfoundland and Labrador's historic cod fisheries attracted local and international fishing fleets for almost five centuries before the Canadian government shut the industry down indefinitely in July 1992. By then, once-plentiful fish stocks had dwindled to near extinction and officials feared they would disappear entirely if the fisheries remained open. The moratorium put about 30,000 people in the province out of work and ended a way of life that had endured for generations in many port communities. It also made evident the vulnerability of marine resources to overexploitation and that existing regulatory regimes were insufficient to protect cod stocks.


Let's isolate the cod stock ***Gadus morhua*** in **East Coast Canada**.
You'll notice that there are multiple different populations, labelled as different stocks.

Read in the three csv files and take a look through:

Use the data in `RAMLDB_v4.66_timeseries.csv` to answer the following questions:
- How did the **total catch** for the different stocks of East Coast Canadian cod vary through time?
- Which of the different stocks had the largest **total catch** in any year?
- When and what was the peak **total catch for all stocks combined**?
- For how many years was the peak catch not **in collapse**, i.e. total catch in a year is greater than 10% of the maximum across all years
- Did the number of fish in the fisheries, defined as **total biomass**, recover after the fishing moratorium in 1992?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy import random
import cartopy.crs as ccrs
```

```python
stock_df = pd.read_csv('data/RAMLDB_v4.66_stock.csv')
ts_df = pd.read_csv('data/RAMLDB_v4.66_timeseries.csv')
tsmetricts_df = pd.read_csv('data/RAMLDB_v4.66_tsmetrics.csv')
```

```python
stock_ids = stock_df[(stock_df['scientificname'].str.contains('Gad', case=True))&(stock_df['region']=='Canada East Coast')]['stockid']
```

```python
cod_data = ts_df[ts_df['stockid'].isin(stock_ids)]
```

```python
cod_data
```

# Dataset cleaning
Now we need to clean the dataset. We have several columns that are not used and rows taht are empty. 

```python
ts_df.describe()
```

```python
cod_ts_df = cod_data[['stockid', 'stocklong', 'year', 'TBbest', 'TCbest', 'ERbest']]
```

Our column names are very specific. We need some notes on what they mean...

```python
tsmetricts_df[tsmetricts_df['tsshort'].isin(['stockid', 'stocklong', 'year', 'TBbest', 'TCbest', 'ERbest'])]
```

```python
cod_ts_df.describe()
```

```python
plt.hist(cod_ts_df['year'])
plt.title('Number or Obs per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
cod_totbiomass = cod_ts_df.groupby(cod_ts_df['year'])['TBbest'].sum()
cod_catch = cod_ts_df.groupby(cod_ts_df['year'])['TCbest'].sum()
cod_mort = cod_ts_df.groupby(cod_ts_df['year'])['ERbest'].mean()

```

```python
fig, ax = plt.subplots(3,1, figsize=(10,8))
ax[0].plot(cod_totbiomass, label= 'Total Biomass', color='blue')
ax[0].set_title('Total Biomass')
ax[0].tick_params(labelbottom=False)

ax[1].plot(cod_catch, label='Catch', color='yellow')
ax[1].set_title('Total Catch')
ax[1].sharex(ax[0])
ax[1].tick_params(labelbottom=False)

ax[2].plot(cod_mort, label='Mortality', color='red')
ax[2].set_title('Average Mortality Rate')
ax[2].sharex(ax[0])

for a in ax:
    a.axvline(x=1992, color='gray', linestyle='--', linewidth=1.5, label='Fisheries Shutdown')
    a.legend()


plt.show
```

Lets see individually, what they look like

```python
plt.figure(figsize=(16,12))
for stock_id in stock_ids:
    plt.plot(cod_data[cod_data['stockid']==stock_id]['year'], cod_data[cod_data['stockid']==stock_id]['TCbest'], label=stock_df[stock_df['stockid']==stock_id]['stocklong'])

plt.axvline(x=1992, color='gray', linestyle='--', linewidth=1.5, label='Fisheries Shutdown')

plt.title('Total Biomass per Stock')
plt.legend(loc='best')
plt.show()
```
