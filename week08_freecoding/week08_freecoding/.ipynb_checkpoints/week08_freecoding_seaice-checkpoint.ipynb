# Explore NASA melting ice sheet data

- Data description: <http://climate.nasa.gov/vital-signs/land-ice/>

The Antarctic and Greenland ice sheet mass loss data is collected by NASA's GRACE satellites. This is a pair of satellites that orbit the Earth together at a distance of about 220 km apart to sense gravity anomalies. When the leading satellite orbits over a part of Earth with slightly stronger gravity, it is pulled ahead a bit faster. Then when it passes the high-gravity area, the satellite slows down again. The distance between the two satellites is consantly measured with a microwave ranging system, and this distance, combined with precise GPS measurements for each satellite in the pair, can be used to generate a map of Earth's gravity field. The melting of ice sheets on land is correlated with a decline in the gravity anomaly over the ice sheet as the ice mass melts and redistributes across the ocean basin.

# Discussion on Annual Pattern and Changes
From Figure 1, we can see that ice extent changes throughout the year. The ice extent for each hemisphere will reach its minimum in the Fall and its maximum in the Spring. The Southern Hemisphere shows a bit more variability throughout the year, but both exhibit the same basic annual pattern.
![Figure 1 of The Ice extent](figures/ice_extent_fullrecord.jpg "Figure 1")

Looking at the full record does not show any clear pattern. So by visualizing all years of the ice record onto one plot (Julian day on the x-axis), we can see a clearer pattern. 

![Figure 2 of the Ice Extent](figures/ice_by_julianday.jpg "Figure 2")

In this figure, we have the Northern Sea Ice extent on top and the Southern Sea Ice extent on the bottom. Each year is visualized by a colour. Early years in the record are darker violets transitioning to bright colours (yellows) in the recent years. From this, we can anecdotally say that the Northern Sea Ice extent is showing a clear trend, with more recent years exhibiting less extent.
```python
# Demonstrating the last 5 years are statistically different. 
# We first need to focus on how we will group the data, then separate the data (last 5-years, and all the other data.)
#.head().tail() # has all the data we want grouped by year. 
n_seaice_yearly.head()
```

```python
# next we will separate them out
# n_seaice_yearly=n_seaice_yearly.reset_index() # need to convert index back to column
n_seaice_yearly_5years = n_seaice_yearly[n_seaice_yearly['Year']>2020]
n_seaice_yearly_full = n_seaice_yearly[n_seaice_yearly['Year']<=2020]
```

```python
# we will first just see how different the averages are different. this will establish our baseline
baseline_diff = np.mean(n_seaice_yearly_5years['extent_mean']) - np.mean(n_seaice_yearly_full['extent_mean'])
baseline_diff
```

We are attempting to prove the last 5 years were different from the rest of the record. By using bootstrapping, we can disprove our null hypothesis. 

*H<sub>0</sub>*: the average of the average yearly extent for the last 5 years is not different from the rest of the temperature record. 

*H<sub>a</sub>*: the average extent of the last 5 years is less than we would expect compared to the previous ice extent. 

```python
# next we will create two random groups from the entire dataset
simulated_diffs = []
num_simulations = 10000
for i in range(num_simulations):
    a = n_seaice_yearly.sample(n=len(n_seaice_yearly_5years), replace=True)
    b = n_seaice_yearly.sample(n=len(n_seaice_yearly_full), replace=True)
    sim_diff = np.mean(a['extent_mean']) - np.mean(b['extent_mean'])
    simulated_diffs.append(sim_diff)
    
```

```python
# our p-value will the number of times that the observed difference was greater than or equal to 
p_value = np.mean(simulated_diffs <= baseline_diff)
p_value
```

```python
plt.figure(figsize=(8,5))
plt.hist(simulated_diffs,alpha=0.5,density=True,color='lightblue', edgecolor='white',label='bootstrap resampled mean differences')
plt.axvline(x=np.percentile(simulated_diffs,2.5),linestyle='--',color='grey',label='95% confidence interval')
plt.axvline(x=np.percentile(simulated_diffs,97.5),linestyle='--',color='grey')
plt.axvline(baseline_diff, label='Observed difference', color='red')
plt.title('Bootstrap distribution of mean Arctic sea ice extent differences \n(1978-2025)', fontsize=14)
plt.xlabel('Difference in mean sea ice extent')
plt.ylabel('Probability Density')
plt.legend(frameon=True, facecolor='white', framealpha=0.8, loc='upper right')
plt.text(baseline_diff, 0.70*plt.ylim()[1], 'Observed difference', rotation=90, color='red', va='top', ha='right')
plt.text(0.70*plt.xlim()[1],0.2,f'P value:{p_value:.2f}')

plt.tight_layout()
plt.savefig('figures/bootstrapping.jpg')
```

```python
# Repeat the process for the southern sea ice data
# next we will separate them out
# s_seaice_yearly=s_seaice_yearly.reset_index() # need to convert index back to column
s_seaice_yearly_5years = s_seaice_yearly[s_seaice_yearly['Year']>2020]
s_seaice_yearly_full = s_seaice_yearly[s_seaice_yearly['Year']<=2020]

# we will first just see how different the averages are different. this will establish our baseline
baseline_diff = np.mean(s_seaice_yearly_5years['extent_mean']) - np.mean(s_seaice_yearly_full['extent_mean'])
print(baseline_diff)
# next we will create two random groups from the entire dataset
simulated_diffs = []
num_simulations = 10000
for i in range(num_simulations):
    a = s_seaice_yearly.sample(n=len(s_seaice_yearly_5years), replace=True)
    b = s_seaice_yearly.sample(n=len(s_seaice_yearly_full), replace=True)
    sim_diff = np.mean(a['extent_mean']) - np.mean(b['extent_mean'])
    simulated_diffs.append(sim_diff)
# our p-value will the number of times that the observed difference was greater than or equal to 
p_value = np.mean(simulated_diffs <= baseline_diff)
print(p_value)
plt.figure(figsize=(8,5))
plt.hist(simulated_diffs,alpha=0.5,density=True,color='lightblue', edgecolor='white',label='bootstrap resampled mean differences')
plt.axvline(x=np.percentile(simulated_diffs,2.5),linestyle='--',color='grey',label='95% confidence interval')
plt.axvline(x=np.percentile(simulated_diffs,97.5),linestyle='--',color='grey')
plt.axvline(baseline_diff, label='Observed difference', color='red')
plt.title('Bootstrap distribution of mean Antarctic sea ice extent differences \n(1978-2025)', fontsize=14)
plt.xlabel('Difference in mean sea ice extent')
plt.ylabel('Probability Density')
plt.legend(frameon=True, facecolor='white', framealpha=0.8, loc='upper right')
plt.text(baseline_diff, 0.70*plt.ylim()[1], 'Observed difference', rotation=90, color='red', va='top', ha='right')
plt.text(0.70*plt.xlim()[1],0.2,f'P value:{p_value:.2f}')

plt.tight_layout()
plt.savefig('figures/bootstrapping_southern.jpg')
```

# Summary
___
From this analysis, we see that sea ice is pretty variable throughout the record and even throughout the year. Annually, we see that northern sea ice reaches its maximum extent in the late winter and reaches its minimum in the fall. Through the record, we see a cyclical pattern across all years from 1978 to the current year (2025). This pattern is driven by the amount of incoming solar radiation and the retention of this heat in the environment; less radiation leads to cooler temperatures, which increase sea ice through the winter season. The summer and fall months receive more energy that can melt sea ice to its annual minimum. Although the general annual pattern of northern sea ice remains the same, the graphs have shifted down, demonstrating the extent has decreased throughout the record. 

Using bootstrapping, we can confidently say that the observed difference between the average ice extent of the last five years and the complete ice record lies outside the 95% bootstrap interval, indicating a significant decline in average ice extent. 

The southern sea ice extent shows a similar annual pattern, but is altered to correspond with its hemisphere's season. Until recently, southern sea ice has not demonstrated the same overall trend as northern sea ice. Differences are highlighted well in this article by [NSICD](https://nsidc.org/learn/ask-scientist/how-does-antarctic-sea-ice-differ-arctic-sea-ice#:~:text=Besides%20occurring%20on%20the%20opposite,growth%20and%20retreat%20for%20decades.). By altering the dates we are interested in (changing from the last 5 to the last 25 years), we can see that the northern sea ice data has exhibited a decline across the record, where the southern sea ice did not.

```python
import pandas as pd
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Aggregate
monthly = n_seaice['Extent 10^6 sq km'].resample('ME').mean()

# Build the Fourier terms for annual (12-month) periodicity
fourier = CalendarFourier(freq='YE', order=3)   # 3 harmonics  6 columns

dp = DeterministicProcess(
    index=monthly.index,
    constant=True,      # intercept
    order=1,            # linear trend
    seasonal=False,     # no dummy months; Fourier handles seasonality
    additional_terms=[fourier],
    drop=False
)

X = dp.in_sample()
y = monthly

# Fit model
model = LinearRegression()
valid = y.notna()
model.fit(X.loc[valid], y.loc[valid])

# Forecast next 10 years (120 months)
X_future = dp.out_of_sample(steps=120)
future_pred = model.predict(X_future)

plt.figure(figsize=(10,5))
plt.plot(y, label='Observed', color='0.5')
plt.plot(y.index, model.predict(X), label='Fitted', color='C0')
plt.plot(X_future.index, future_pred, '--', label='10-year forecast', color='C1')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Mean monthly sea-ice extent (million km)')
plt.title('Fourier + Linear trend model on monthly sea-ice extent')
plt.tight_layout()
plt.show()
```
