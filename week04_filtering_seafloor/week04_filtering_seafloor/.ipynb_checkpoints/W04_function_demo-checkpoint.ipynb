```python
def doSomething(deg, radius, color="red"):
    x=deg*radius
    print(color)
    return x

answer = doSomething( 30, 6371, "pink!")
```

```python
answer
```

```python
def doRedThing(deg, radius, color="red"):
    x=deg*radius
    print(f"color is {color}")
    return x

answer = doSomething( 30, 6371, "pink!")
print(f"did pink and answer={answer}")

green_ans = doRedThing(1, 2, "green")
print(f"red with green ans = {green_ans}")
red_ans = doRedThing(1, 3)
print(f"red ans = {red_ans}")
```
